```python
import torch
import gensim
import seaborn as sns
import numpy as np
import snntorch as snn

from tqdm import tqdm
from torchtext.data import utils
from matplotlib import pyplot as plt
```

```python
DEVICE = (f'cuda:0' if torch.cuda.is_available() else 'cpu')
print(f'Using device {DEVICE}')
if torch.cuda.is_available():
    print(torch.cuda.get_device_name())
```

```python
# Path to the saved word embeddings file
EMBED_KEYS_PATH = 'abstracts200_trained_normalized.wordvectors'
# Unique identifier for null (for padding)
NULL_WORD = '\0'
# Path to trained SNN model
model_path = 'snn_model-6_v1.pth'
# Number of time steps
T = 45
# membrane decay constant
beta = 1.0
```

```python
wv = gensim.models.KeyedVectors.load(EMBED_KEYS_PATH, mmap='r') # Get the saved word vectors matrix
```

```python
tk = utils.get_tokenizer('spacy')
null_word = '\0'
d = AbstractDataset('CleanedAVdata.csv', 'Abstract', 'IPCR Classifications', tk, wv.key_to_index,
                    null_word=null_word, min_len=30, verbose=True)
```

```python
model = AbstractSNN_1(T, EMBED_KEYS_PATH, null_word=NULL_WORD, beta=beta).to(DEVICE)
model = load_model(model, model_path)
# Save and detach the model's inhibitor matrix for ease of access
inhibitor = model.inhibitor.detach().cpu().data
```

## 1) Get the the words best associated with each topic 
Find the words that elicit the most spikes for each topic

```python
key_to_ind = wv.key_to_index
spikes = []
n_chunks = 40

data = torch.empty(len(key_to_ind), 669, dtype=torch.int32)
for i, word in list(enumerate(key_to_ind.keys())):
    x = torch.tensor(key_to_ind[word]).repeat(669)
    data[i] = x

data = data.to(DEVICE)

model.eval()
with torch.no_grad():
    for xbatch in tqdm(torch.chunk(data, n_chunks)):
        total_spk = forward_pass(model, T, xbatch).squeeze().sum(dim=0).cpu()
        spikes.append(total_spk)

spikes = torch.cat(spikes, dim=0)
```

```python
s1 = sns.heatmap(spikes, cbar_kws={'label': 'Number of spikes'})
s1.set(xlabel='Topics', ylabel='Words')
```

## 3) Visualize the difference between how the inhibition for a particular topic differs from the average inhibition when compared to commonly co-occurring topics
For each topic i, we first find all topics that i co-occurs with more than 10% of the time. Then, we find the inhibitory currents that topic i sends to these co-occurring topics, and subtract the average inhibitory currents that all topics send to these particular co-occurring topics. 

We expect most values to be negative, which means that when topic i co-occurs more than 10% of the time with another topic, it typically sends less inhibitory current to those topics. 

```python
def get_mean_inhib_diff(co_occ_threshold, average='mean', plot=False):
    mean_inhib_send = (inhibitor.mean(dim=0) if average == 'mean' else inhibitor.median(dim=0)[0])
    if plot:
        fig, ax = plt.subplots()

    inhib_diff = []
    n_neg, n_total = 0, 0
    for i in range(91):
        # Topics that co-occur with topic i more than 10% of the time
        ind = d.labels[(d.labels[:,i] == 1)].mean(dim=0) > co_occ_threshold
        # The difference between the inhibition values that topic i sends, and the average inhibition values across all topics
        diff = inhibitor[i][ind] - mean_inhib_send[ind]
        inhib_diff.append(diff)
        n_neg += (diff < 0).sum()
        n_total += len(diff)

    # inhib_diff = torch.cat(inhib_diff, dim=1)
    for i, diff in enumerate(inhib_diff):
        data = np.array([[i]*len(diff), diff.numpy()]).T
        if plot:
            ax.scatter([i]*len(diff), diff.numpy(), color='orange')
    if plot:
        ax.axhline(0, color='blue')
        ax.set_xlabel('Topic')
        ax.set_ylabel('Difference in inhibition weight')
        fig.suptitle('Difference between Topic i\'s inhibition and average inhibition across all topics ')
    
    return (n_neg/n_total).item()
    
```

```python
get_mean_inhib_diff(-1, plot=True)
```

```python
perc_neg_diff_mean = [get_mean_inhib_diff(thresh, average='mean')*100 for thresh in np.arange(0, 1, 0.01)]
perc_neg_diff_median = [get_mean_inhib_diff(thresh, average='median')*100 for thresh in np.arange(0, 1, 0.01)]

fig, ax = plt.subplots()

ax.plot(perc_neg_diff_mean, label='mean')
ax.plot(perc_neg_diff_median, label='median')
fig.suptitle('Percent of lower-than-average inhibition vs. co-occurrence threshold')
ax.set_xlabel('Co-occurrence percentage threshold')
ax.set_ylabel('Percentage of negative difference')
plt.legend()
```
